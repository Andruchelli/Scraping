```python
filtred_child = [] # добавляем пустой массив, куда будем записывать всех полученных "детей"

for child in first_block.children:
    if child == "\n":
        continue
    filtred_child.append(child)
```

```python
len(filtred_child) # считаем количество полученных результатов
```

```python
img = img_1.get_attribute_list('src')[0] # записываем в переменную img ссылку на изображение
img
```

```python
price = filtred_child[2].find('div', attrs = {"class": "product__price-new"}).text.strip() # записываем в переменную price стоимость стула
price
```

```python
url_template = "https://designboom.ru/catalog/mebel/stulya/={}"

base = []

for i in range(1, 6):
    url_resolved = url_template.format(i)
    response = r.get(url_resolved)
    
    assert response.ok, "Some error"
    
    soup = BeautifulSoup(response.text, "lxml")
    blocks = soup.find_all('div', attrs = {"class": "product__container"})
    
    for block in blocks:
        filtered_children = []
        item = {} # записываем базу

        for child in block.children:
            if child == "\n":
                continue
            filtered_children.append(child)

        div = filtered_children[0]
        img = img_1.get_attribute_list('src')[0]
        item["img"] = "https://designboom.ru" + img # ссылка на изображение
        item["name"] = filtred_child[2].find('p', attrs = {"class": "product__title mb-0"}).text.strip() # название стула

        #price_block = filtered_children[2]
        #if price_block.findAll("span", attrs={"class": "product_small_old_price"}):
            #old_price_block = price_block.findAll("span", attrs={"class": "product_small_old_price"})[0] 
            #item["old_price"] = old_price_block.text
        item["price"] = filtred_child[2].find('div', attrs = {"class": "product__price-new"}).text.strip() # цена стула


        base.append(item)
```
