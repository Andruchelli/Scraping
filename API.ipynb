## Простой API, Get запрос

```python
todo_api_url = "https://jsonplaceholder.typicode.com/todos/1" # сохраняем ссылку на сайт в переменную todo_api_url
```

```python
requests.get(todo_api_url) # вызываем функцию get из библиотеки requests
```

```python
r = requests.get(todo_api_url) # сохраняем ответ в переменную r
```

```python
# посмотрим какие есть методы
r # после переменной r стави точку (.) и нажимаем tab, чтобы появился список возможных методов переменной
```

```python
# прошло ли все успешно
r.ok
```

```python
# код возврата
r.status_code
```

```python
# заголовки, которые отправил сервер, метаинформация про ответ
r.headers
```

```python
# заголовки запроса
r.request.headers
```

```python
# тело ответа
r.content
```

## Get запрос с параметрами

```python
r1 = requests.get("https://newsapi.org/v2/top-headlines?country=fr&apiKey=36c7104302364b64a945e7e35eb7d524")
r1.json()
```

```python
api_key = "36c7104302364b64a945e7e35eb7d524"
base_url = "https://newsapi.org/v2/everything" # базовая страница сайта без дополнительных параметров
```

```python
query = "bitcoin price"

resp = requests.get(base_url, params={
    "apiKey": api_key, "q": query, "language": "en"
})
```

```python
resp.ok
```

```python
body = resp.json()
```

```python
body
```

```python
#body["articles"]
```

```python
for article in body["articles"]:
    print(article["title"], article["url"], end="\n\n") # выведем только title и url полученных заголовков; end="\n\n" - добавляем 2 возврата каретки для удобного чтения полученной информации 
```
