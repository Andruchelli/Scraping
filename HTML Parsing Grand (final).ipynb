## Постановка задачи:

спарсить сайт с ценами https://www.grandfs.ru/catalog/krovati/ в текстовый файл, в котором будут храниться:
* ссылка на изображение товара
* его название
* цена
* ссылка на карточку товара

```python
import requests as r
import re
from bs4 import BeautifulSoup
```

```python
url_template = "https://www.grandfs.ru/catalog/krovati/"

base = []

response = r.get(url_template)

soup = BeautifulSoup(response.text, "lxml")

# ищем все строки на странице с тегом div и параметрами (attrs), где находится информация о названии товара и ссылка на него
blocks = soup.find_all('div', attrs = {"class": "item"})

for block in blocks:
    filtered_child = [] # добавляем пустой массив, куда будем записывать всех полученных "детей"
    item = {} # записываем базу
    
    for child in block.children:
        if child == "\n":
            continue
        filtered_child.append(child)
        
    brand = filtered_child[0].text.strip() # сохраняем в переменную brand название бренда
    title = filtered_child[1].text.strip() # сохраняем в переменную title название товара
    img = filtered_child[2].find("div", attrs = {"class": "item__img"}).get("style")
    img = re.search(r'\((.+?)\)', img)[1] # с помощью регулярных выражений выводим только ту часть, которая находится в скобках
    link = filtered_child[1].get_attribute_list('href')[0] # получаем ссылку на карточку товара
    full_link = "https://www.grandfs.ru" + link
  
    item["name"] = ' '.join([brand, title]) # получаем полное название товара с названием бренда
    item["img"] = "https://www.grandfs.ru" + img # полная ссылка на изображение товара
    item["link"] = "https://www.grandfs.ru" + link # полная ссылка на товар
    item["old_price"] = filtered_child[3].find("div", attrs = {"class": "item__oldprice"}).text.strip() # получаем старую цену товара
    item["new_price"] = filtered_child[3].find("div", attrs = {"class": "item__newprice"}).text.strip() # получаем новую цену товара

    base.append(item)
```
