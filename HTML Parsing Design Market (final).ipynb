```python
import requests as r
from bs4 import BeautifulSoup
```

```python
response = r.get("https://designboom.ru/catalog/mebel/stulya")
```

```python
response.status_code
```

```python
url_template = "https://designboom.ru/catalog/mebel/stulya/?PAGEN_1={}"

base = []

for i in range(1, 6):
    url_resolved = url_template.format(i)
    response = r.get(url_resolved)
    
    assert response.ok, "Some error" # цикл не сработает, если выпадет ошибка

    soup = BeautifulSoup(response.text, "lxml")
    blocks = soup.find_all('div', attrs = {"class": "product__container"})

    for block in blocks:
        filtered_child = []
        item = {} # записываем базу

        for child in block.children:
            if child == "\n":
                continue
            filtered_child.append(child)

        div = filtered_child[1]
        img_1 = div.findAll('img')[0]
        img = img_1.get_attribute_list('src')[0]
        info = filtered_child[2]
        item["img"] = "https://designboom.ru" + img # ссылка на изображение
        item["name"] = info.find('p', attrs = {"class": "product__title mb-0"}).text.strip()
        item["price"] = info.find('div', attrs = {"class": "product__price-new"}).text.strip()

        base.append(item)
```
