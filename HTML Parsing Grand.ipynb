## Постановка задачи:

спарсить сайт с ценами https://www.grandfs.ru/catalog/krovati/ в текстовый файл, в котором будут храниться:
* ссылка на изображение товара
* его название
* цена
* ссылка на карточку товара

```python
import requests as r
import re
from bs4 import BeautifulSoup
```

```python
response = r.get("https://www.grandfs.ru/catalog/krovati/")
```

```python
response.status_code # получаем код статуса выполнения обращения к странице сайта
```

```python
response.content # получаем html код запрашиваемой страницы сайта
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
soup.find_all('a') # ищем все строки на странице с тегом a
```

```python
# ищем все строки на странице с тегом div и параметрами (attrs), где находится информация о названии товара и ссылка на него
blocks = soup.find_all('div', attrs = {"class": "item"})
blocks
```

```python
first_block = blocks[0] # записываем в переменную first_block код с информацией о первом товаре
```

```python
list(first_block.children) # возвращаем "детей" блока
```

```python
# посчитаем количество "детей" в полученном списке
len(list(first_block.children))
```

```python
filtered_child = [] # добавляем пустой массив, куда будем записывать всех полученных "детей"

for child in first_block.children:
    if child == "\n":
        continue
    filtered_child.append(child)
```

```python
len(filtered_child) # считаем количество полученных результатов
```

```python
a = filtered_child[0] # записываем в переменную a первого "ребёнка"
a
```

```python
brand = a.text.strip() # сохраняем в переменную brand название бренда
brand
```

```python
b = filtered_child[1]
b
```

```python
title = b.text.strip() # сохраняем в переменную title название товара
title
```

```python
name = ' '.join([brand, title]) # получаем полное название товара с названием бренда
name
```

```python
img_1 = filtered_child[2] # записываем в переменную img_1 третьего "ребёнка"
img_1
```

```python
img = filtered_child[2].find("div", attrs = {"class": "item__img"}).get("style")
img = re.search(r'\((.+?)\)', img)[1] # с помощью регулярных выражений выводим только ту часть, которая находится в скобках
full_img = "https://www.grandfs.ru" + img # генерируем полную ссылку на товар
full_img
```

```python
link = filtered_child[1].get_attribute_list('href')[0] # получаем ссылку на карточку товара
full_link = "https://www.grandfs.ru" + link
full_link
```

```python
old_price = filtered_child[3].find("div", attrs = {"class": "item__oldprice"}).text.strip() # получаем старую цену товара
old_price
```

```python
new_price = filtered_child[3].find("div", attrs = {"class": "item__newprice"}).text.strip() # получаем новую цену товара
new_price
```
